```python
mapping = {
    "PL COTA MZ": "11703",
    "PL COTA SR": "11701",
    "DIREITOS CREDITORIOS - A VENCER": "12990",
    "DIREITOS CREDITORIOS - VENCIDOS": "13190",
    "COTAS F.I (RENDA FIXA)": "15805",
    "PDD": "11902",
    "CAIXA":"10108",
    "TAXAS DIFERIDAS CVM": "19701",
    "TAXAS DIFERIDAS ANBIMA": "19703",
    "TAXAS ADM": "23702",
    "TAXAS GESTAO": "23711",
    "TAXAS AUDITORIA": "23716",
    "TAXAS CUSTODIA": "23719",
    "SELIC": "23706",
    "CETIP": "23704"
}

nomes = {
    "AGROFORTE FIDC - SUBORDINADA": "FIDC AGROFORTE",
    "APG FIDC - SUBORDINADA": "FIDC APG"
}



def ajustar_valor(valor):
    try:
        # Verifica se o valor é uma string e contém números antes de tentar manipulá-lo
        if isinstance(valor, str) and valor.isdigit():
            valor = valor[:-2] + ',' + valor[-2:]  # Insere a vírgula antes dos dois últimos caracteres
        elif isinstance(valor, str) and ',' in valor and '.' not in valor:
            valor = valor.replace(',', '.')  # Converte para formato decimal caso esteja usando vírgula
        elif isinstance(valor, str) and '.' in valor and ',' in valor:
            valor = valor.replace('.', '').replace(',', '.')  # Trata casos com separadores de milhar e decimal
        return valor
    except Exception as e:
        print(f"Erro ao ajustar valor '{valor}': {e}")
        return valor  # Retorna o valor original em caso de erro


def ajustar_quantidade(valor):
    try:
        # Verifica se o valor é None
        if valor is None:
            return valor  # Se for None, retorna o valor sem alteração

        # Verifica se o valor contém vírgula
        if ',' not in str(valor):
            # Converte o valor para string, coloca a vírgula antes dos últimos 8 dígitos
            valor_str = str(valor)
            # Insere a vírgula antes dos últimos 8 caracteres
            valor_ajustado = valor_str[:-8] + ',' + valor_str[-8:]
            return valor_ajustado
        
        return valor  # Retorna o valor original se já contiver vírgula
    except Exception as e:
        print(f"Erro ao ajustar quantidade '{valor}': {e}")
        return valor  # Retorna o valor original em caso de erro


   


# Lista todos os arquivos na pasta de entrada com extensão .XLS
files = [f for f in os.listdir(input_directory) if f.endswith('.XLS')]

# Lista para armazenar os DataFrames processados
dataframes_processados = []

for file in files:
    file_path = os.path.join(input_directory, file)
    print(f"Processando arquivo: {file}")

    try:
        tables = pd.read_html(file_path)  
        if len(tables) == 0:
            print(f"Nenhuma tabela encontrada no arquivo '{file}'")
            continue
        df = pd.concat(tables, ignore_index=True)
    except Exception as e:
        print(f"Erro ao processar arquivo '{file_path}': {e}")
        print(traceback.format_exc())
        continue

    try:
        dfs = [df.dropna(how='all', axis=0).dropna(how='all', axis=1).reset_index(drop=True).iloc[:, :18] for df in tables]

        dfs_filtrados = []
        for df in dfs:
            if df.iloc[0, 0] in tabelas:
                df['Tabela'] = df.iloc[0, 0]
                dfs_filtrados.append(df)

        df = pd.concat(dfs_filtrados, ignore_index=True)
        df.columns = range(len(df.columns))

        data = df[df[df.columns[0]] == 'Data de Posição:'].iloc[0, 1]
        fundos = df[df[df.columns[0]] == 'Cliente:'].iloc[:, :2]

         # Verificação adicional para Data de Posição
        data_rows = df[df[df.columns[0]] == 'Data de Posição:']
        if data_rows.empty:
            print(f"Erro: 'Data de Posição:' não encontrada no arquivo {file}")
            continue
        
        data = data_rows.iloc[0, 1]
        print(f"Data de Posição encontrada no arquivo {file}: {data}")


        # print(df[df[0] == 'SOCDILP'].iloc[0,4])

        df['NmFundo'] = None

        for i in range(len(fundos)):
            nome_fundo = fundos.iloc[i, 1]
            idx_inicial = fundos.index[i]

            if i + 1 < len(fundos):
                idx_final = fundos.index[i + 1] - 1
            else:
                idx_final = df.index[-1]

            df.loc[idx_inicial:idx_final, 'NmFundo'] = nome_fundo

        

        df['DtPosicao'] = data
       

        # Filtros e transformações
        df = df[df['NmFundo'].str.contains('SUBORDINADA', na=False)]
        df = df[df[2].isin(['Tesouraria', 'Conta Corrente', 'Fundos de Investimento - Outros Fundos', 'Outros Ativos', 'Renda Fixa', 'Contas a Pagar/Receber'])]
        df = df[(df[0].isin(['SOCDILP', 'TOTAL', 'SUBTOTAL', 'MEZAN', 'SRP', 'AGROFAVE', 'AGROFVEN', 'APGAV', 'APGVE'])) | (df[2] == 'Contas a Pagar/Receber')]
        df = df.drop(df[(df[0] == 'TOTAL') & (df[2] == 'Fundos de Investimento - Outros Fundos')].index)
        df = df.drop(df[(df[0] == 'TOTAL') & (df[2] == 'Renda Fixa')].index)
        df = df.drop(df[(((df[0] == 'TOTAL') | (df[0] == 'Descrição')) & (df[2] == 'Contas a Pagar/Receber')) | (df[0] == 'Contas a Pagar/Receber')].index)
        df = df.drop_duplicates(subset=None, keep='last', inplace=False).reset_index(drop=True)

        colunas_para_manter = [0, 1, 2, 3, 4, 10, 11, 16, 'NmFundo', 'DtPosicao']
        df = df[colunas_para_manter]

        df.loc[df[0] == 'SUBTOTAL', 0] = df[0].shift(1)
        df = df.drop(df[((df[0] == 'MEZAN') | (df[0] == 'SRP')) & (pd.isna(df[16]))].index)
        
        df['Valor'] = df.apply(
            lambda row: pd.to_numeric(row[1].replace('.', '').replace(',', '.'), errors='coerce') / 100 if row[2] == 'Tesouraria' 
            else row[4] if row[2] == 'Conta Corrente'
            else row[10] if row[2] == 'Fundos de Investimento - Outros Fundos'
            else row[3] if row[2] == 'Outros Ativos'
            else row[16] if row[2] == 'Renda Fixa'
            else row[1] if row[2] == 'Contas a Pagar/Receber'
            else None, 
            axis=1
        )

        df['Quantidade'] = df.apply(
            lambda row: row[4] if row[2] == 'Fundos de Investimento - Outros Fundos' 
            else row[11] if row[2] == 'Renda Fixa'
            else None, 
            axis=1
        )

        for fundo in df['NmFundo'].unique():
            tesouraria = df[(df[2] == 'Tesouraria') & (df['NmFundo'] == fundo)]
            conta_corrente = df[(df[2] == 'Conta Corrente') & (df['NmFundo'] == fundo)]
        
            if len(tesouraria) == 1 and len(conta_corrente) == 1:
                tesouraria_valor = pd.to_numeric(tesouraria['Valor'].iloc[0], errors='coerce')
                conta_corrente_valor = pd.to_numeric(conta_corrente['Valor'].iloc[0], errors='coerce')
                df.at[tesouraria.index[0], 'Valor'] = tesouraria_valor + conta_corrente_valor
                df.drop(conta_corrente.index, inplace=True)

        def preencher_valores(row):
            if row[2] == 'Tesouraria':
                return 'SALDO DE CAIXA', 'SALDO DE CAIXA', 'CAIXA'
            elif isinstance(row[0], str) and 'av' in row[0].lower() and row[2] == 'Fundos de Investimento - Outros Fundos':
                return 'A VENCER', 'DIREITOS CREDITORIOS', 'DIREITOS CREDITORIOS - A VENCER'
            elif isinstance(row[0], str) and 've' in row[0].lower() and row[2] == 'Fundos de Investimento - Outros Fundos':
                return 'VENCIDOS', 'DIREITOS CREDITORIOS', 'DIREITOS CREDITORIOS - VENCIDOS'
            elif row[2] == 'Fundos de Investimento - Outros Fundos':
                return row[1], 'RENDA FIXA', 'COTAS F.I (RENDA FIXA)'
            elif row[2] == 'Outros Ativos':
                return 'PDD', 'PDD', 'PDD'
            elif row[2] == 'Renda Fixa':
                return row[0], 'PL Cotas', row[0]
            elif row[2] == 'Contas a Pagar/Receber':
                return row[0], row[2], 'OUTROS'
            else:
                return 'OUTROS', 'OUTROS', 'OUTROS'
        
        df[['Descricao', 'Grupo', 'Categoria']] = df.apply(lambda row: preencher_valores(row), axis=1, result_type='expand')
        df['Categoria'] = df['Categoria'].replace({'MEZAN': 'PL COTA MZ', 'SRP': 'PL COTA SR'})

        colunas_para_manter = ['NmFundo', 'DtPosicao', 'Valor', 'Quantidade', 'Categoria', 'Grupo', 'Descricao']
        df = df[colunas_para_manter]


        # Converta a coluna para string e trate NaN
        df["Valor"] = df["Valor"].fillna("").astype(str)

        df['Descricao'] = df['Descricao'].str.upper()

        def categorizar(row):
            if (row['Grupo'] == 'Contas a Pagar/Receber' and 
                not pd.Series(row['Valor']).str.contains(r"\(").any() and 
                "DIFERIMENTO" not in row['Descricao']):
                row['Grupo'] = 'VALORES A RECEBER'
                row['Categoria'] = 'VALORES A RECEBER'

            elif (row['Grupo'] == 'Contas a Pagar/Receber' and
                 "CVM" in row["Descricao"] and
                 "DIFERIMENTO" in row["Descricao"]):
                row['Grupo'] = 'TAXAS DIFERIDAS'
                row['Categoria'] = 'TAXAS DIFERIDAS CVM'

            elif (row['Grupo'] == 'Contas a Pagar/Receber' and
                 "ANBIMA" in row["Descricao"] and
                 "DIFERIMENTO" in row["Descricao"]):
                row['Grupo'] = 'TAXAS DIFERIDAS'
                row['Categoria'] = 'TAXAS DIFERIDAS ANBIMA'

            elif (row['Grupo'] == 'Contas a Pagar/Receber' and
                 "ADMINISTRAÇÃO" in row["Descricao"] and
                 "TAXA" in row["Descricao"]):
                row['Grupo'] = 'DESPESAS'
                row['Categoria'] = 'TAXAS ADM'

            elif (row['Grupo'] == 'Contas a Pagar/Receber' and
                 "GESTÃO" in row["Descricao"] and
                 "TAXA" in row["Descricao"]):
                row['Grupo'] = 'DESPESAS'
                row['Categoria'] = 'TAXAS GESTAO'

            elif (row['Grupo'] == 'Contas a Pagar/Receber' and
                 "AUDITORIA" in row["Descricao"] and
                 "TAXA" in row["Descricao"]):
                row['Grupo'] = 'DESPESAS'
                row['Categoria'] = 'TAXAS AUDITORIA'

            elif (row['Grupo'] == 'Contas a Pagar/Receber' and
                 "CUSTÓDIA" in row["Descricao"] and
                 "TAXA" in row["Descricao"]):
                row['Grupo'] = 'DESPESAS'
                row['Categoria'] = 'TAXAS CUSTODIA'

            elif row['Grupo'] == 'Contas a Pagar/Receber':
                row['Grupo'] = 'DESPESAS'
                row['Categoria'] = 'TAXAS OUTROS'

            return row

        df = df.apply(categorizar, axis=1)

        df["Cod"] = df["Categoria"].map(mapping)
        df['NmFundo'] = df['NmFundo'].map(nomes)
        
        df['Valor'] = df['Valor'].astype(str).apply(
            lambda x: x.replace('.', ',') if '.' in x and x.replace('.', '').isdigit() else x
        )

        df['Valor'] = df['Valor'].str.replace('(', '-', regex=False)
        df['Valor'] = df['Valor'].str.replace(')', '', regex=False)

        df['Quantidade'] = df['Quantidade'].str.replace('(', '-', regex=False)
        df['Quantidade'] = df['Quantidade'].str.replace(')', '', regex=False)

        df['Valor'] = df['Valor'].apply(ajustar_valor)
        df['Quantidade'] = df['Quantidade'].apply(ajustar_quantidade)

        # print(f"Arquivo '{file}' processado com sucesso.")

        df['Descricao'] = df['Descricao'].replace(mapping)

        dataframes_processados.append(df)
        print(f"Arquivo '{file}' processado com sucesso.")
        
    except Exception as e:
        print(f"Erro durante processamento do arquivo {file}: {e}")
        print(traceback.format_exc())


# Concatena todos os DataFrames em um único DataFrame final
df_final = pd.concat(dataframes_processados, ignore_index=True)

#Pegando o tipo de fundo a partir de NmFundo
df['NmFundo'] = df['NmFundo'].str.strip()  
df_final['Tp_Fundo'] = 'FIDC'


#Pega a data de venciento
df_final['DataVenc'] = None

#Pega a Data de aplicação
df_final['DataAplic'] = None

# Renomeando as colunas
df_final = df_final.rename(columns={
    'Quantidade': 'Qnt',
    'Valor': 'VlrMercado' 
})

df_final['Custodiante'] = 'Singulare'

# Definir a nova ordem das colunas
nova_ordem = ['NmFundo', 'Tp_Fundo', 'DtPosicao', 'Descricao', 'Qnt', 'VlrMercado', 'DataVenc', 'DataAplic', 'Categoria', 'Cod', 'Grupo', 'Custodiante']

# Reorganizar as colunas no DataFrame
df_final = df_final[nova_ordem]

# Ajustar a coluna 'DtPosicao' para o formato ano-mês-dia (YYYY-MM-DD)
df_final['DtPosicao'] = df_final['DtPosicao'] = pd.to_datetime(df_final['DtPosicao'], dayfirst=True, errors='coerce').dt.strftime('%Y-%m-%d')


# Exibir as primeiras linhas do DataFrame ajustado
df_final.head(60)
```

```python
import os
import pandas as pd
import chardet

# Dicionário grupo_map (exemplo, por favor ajuste conforme necessário)
grupo_map = {
    '02-Fundos-Fundos': 'PORTFOLIO INVESTIDO',
    '01-Renda-Fixa': 'RENDA FIXA',
    '06-Patrimonio-Totais': 'SALDO DE CAIXA',
    '03-Outros-Ativos': 'PDD'
}

grupo_cod = {
    'PORTFOLIO INVESTIDO': 15804,
    'RENDA FIXA': 15805,
    'DESPESAS': 'OUTROS',
    'SALDO DE CAIXA': 10108,
    'ACOES': 'OUTROS',
    'PDD': "PDD/MTM"
}

descricao_dict2 = {
    "VALORES RECEBER": "VALORES RECEBER",
    "FUNDOS DE DIREITO CREDITÓRIO": "FUNDOS DE DIREITO CREDITÓRIO",
    "FUNDOS DE RENDA FIXA": "FUNDOS DE RENDA FIXA",
    "COTAS DE FDO DIR. CREDITÓRIO": "COTAS DE FDO DIR. CREDITÓRIO",
    "TAXA CVM DIFERIDO": "TAXA CVM DIFERIDO",
    "TAXA ANBID DIFERIDO": "TAXA ANBID DIFERIDO",
    "PORTFOLIO INVESTIDO": "PORTFOLIO INVESTIDO",
    "DESPESAS": "DESPESAS",
    "TAXA DE ADMINISTRACAO": "TAXA DE ADMINISTRACAO",
    "CETIP": "CETIP",
    "RENDA FIXA": "RENDA FIXA",
    "TAXA DE GESTÃO": "TAXA DE GESTÃO",
    "BONUS DE PERFORMANCE": "BONUS DE PERFORMANCE",
    "PATRIMONIO LIQUIDO": "PATRIMONIO LIQUIDO",
    "SALDO DE CAIXA": "SALDO DE CAIXA",
    "PDD": "PDD",
    "ACOES": "ACOES",
    "DEMAIS DIREITOS CREDITÓRIOS": "DEMAIS DIREITOS CREDITÓRIOS",
    "TAXA DE CUSTÓDIA": "TAXA DE CUSTÓDIA",
    "AUDITORIA": "AUDITORIA",
    "TAXA DE CONTROLADORIA": "TAXA DE CONTROLADORIA",
    "DIREITOS CREDIT. A IDENTIFICAR": "DIREITOS CREDIT. A IDENTIFICAR",
    "CP AJUSTE COTA DEVEDOR": "CP AJUSTE COTA DEVEDOR",
    "SELIC": "SELIC",
    "IR": "IR",
    "TAXA CVM": "TAXA CVM",
    "TAXA DE CONSULTORIA": "TAXA DE CONSULTORIA",
    "PL COTA SÊNIOR": "PL COTA SÊNIOR",
    "VALORES RECEBER - FUNDOS": "VALORES RECEBER - FUNDOS",
    "DC. S. RET. DE RISCOS A VENCER": "DC. S. RET. DE RISCOS A VENCER",
    "VALORES A PAGAR": "VALORES A PAGAR",
    "PL - COTA MEZANINO": "PL - COTA MEZANINO",
    "DESPESAS ANTECIPADAS A DIFERIR": "DESPESAS ANTECIPADAS A DIFERIR",
    "PL - COTA DE CLASSE": "PL - COTA DE CLASSE",
    "AJUSTE DE COTAS ATIVO": "AJUSTE DE COTAS ATIVO",
    "OUTRAS DESPESAS": "OUTRAS DESPESAS",
    "TAXA REBATE": "TAXA REBATE",
    "DESPESAS BANCÁRIAS": "DESPESAS BANCÁRIAS",
    "FUNDOS REFERENCIADO": "FUNDOS REFERENCIADO",
    "COTAS A RESGATAR": "COTAS A RESGATAR",
    "VALORES A INTEGRALIZAR": "VALORES A INTEGRALIZAR",
    "SERVIÇO DE AGENTE DE COBRANÇA": "SERVIÇO DE AGENTE DE COBRANÇA",
    "SERVIÇOS DE ESTRUTURAÇÃO": "SERVIÇOS DE ESTRUTURAÇÃO",
    "COTAS A EMITIR": "COTAS A EMITIR"
}

#Função patra atualizar Grupo como renda fixa
def atualizar_grupo(row):
    if row['Grupo'] == 'PDD':
        row['Descricao'] = 'PDD'  # Modificar diretamente a classificação
    return row

#Função patra atualizar descrição
def atualizar_desc(row):
    if row['Grupo'] == 'SALDO DE CAIXA':
        row['Descricao'] = 'SALDO DE CAIXA'  # Modificar diretamente a classificação
    return row

def ajustar_valor(valor):
    try:
        # Verifica se o valor é uma string e contém números antes de tentar manipulá-lo
        if isinstance(valor, str) and valor.isdigit():
            valor = valor[:-2] + ',' + valor[-2:]  # Insere a vírgula antes dos dois últimos caracteres
        elif isinstance(valor, str) and ',' in valor and '.' not in valor:
            valor = valor.replace(',', '.')  # Converte para formato decimal caso esteja usando vírgula
        elif isinstance(valor, str) and '.' in valor and ',' in valor:
            valor = valor.replace('.', '').replace(',', '.')  # Trata casos com separadores de milhar e decimal
        return valor
    except Exception as e:
        print(f"Erro ao ajustar valor '{valor}': {e}")
        return valor  # Retorna o valor original em caso de erro



# Caminhos das pastas onde estão os arquivos
pastas_arquivos = [
    '//172.31.8.209/Files/CataliseInvestimentos/14. BigData Catalise/Staging/Carteira Diaria/Master Diaria/06-Patrimonio-Totais',
    '//172.31.8.209/Files/CataliseInvestimentos/14. BigData Catalise/Staging/Carteira Diaria/Master Diaria/01- Renda Fixa',
    '//172.31.8.209/Files/CataliseInvestimentos/14. BigData Catalise/Staging/Carteira Diaria/Master Diaria/02-Fundos-Fundos',
    '//172.31.8.209/Files/CataliseInvestimentos/14. BigData Catalise/Staging/Carteira Diaria/Master Diaria/03-Outros_Ativos'
]

# Inicializar DataFrames individuais
df_06_patrimonio = pd.DataFrame()
df_01_renda_fixa = pd.DataFrame()
df_02_fundos = pd.DataFrame()
df_outros_ativos = pd.DataFrame()

# Função para detectar a codificação e carregar os dados
def carregar_csv(caminho_arquivo):
    with open(caminho_arquivo, 'rb') as f:
        result = chardet.detect(f.read())
    encoding_detected = result['encoding']
    return pd.read_csv(caminho_arquivo, encoding=encoding_detected, delimiter=";", on_bad_lines='skip')

# Processar cada pasta
for pasta in pastas_arquivos:
    for arquivo in os.listdir(pasta):
        if arquivo.endswith('.csv'):
            caminho_arquivo = os.path.join(pasta, arquivo)
            print(f"Processando: {arquivo} na pasta {pasta}")
            
            df = carregar_csv(caminho_arquivo)
            
            if '06-Patrimonio-Totais' in pasta:
                colunas_desejadas = ['CARTEIRA', 'DATA', 'SALDOCAIXAATUAL']
                if all(col in df.columns for col in colunas_desejadas):
                    df_filtrado = df[colunas_desejadas]
                    df_filtrado.rename(columns={
                        'CARTEIRA': 'NmFundo',
                        'DATA': 'DtPosicao',
                        'SALDOCAIXAATUAL': 'VlrMercado'
                    }, inplace=True)
                    df_filtrado['Descricao'] = None
                    df_filtrado['Qnt'] = None
                    df_06_patrimonio = pd.concat([df_06_patrimonio, df_filtrado], ignore_index=True)
                 
            
            elif '01- Renda Fixa' in pasta:
                colunas_relevantes = [
                    'CARTEIRA', 'DATA', 'TITULO', 'VALORLIQUIDO',
                    'DATACOMPRA', 'DATAVENCIMENTO', 'QUANTIDADE'
                ]
                if all(col in df.columns for col in colunas_relevantes):
                    df_filtrado = df[colunas_relevantes]
                    df_filtrado = df_filtrado[
                        df_filtrado['TITULO'].str.contains(r'COMP|NTN-B|CRI', na=False, case=False)
                    ]
                    df_filtrado.rename(columns={
                        'CARTEIRA': 'NmFundo',
                        'DATA': 'DtPosicao',
                        'VALORLIQUIDO': 'VlrMercado',
                        'DATACOMPRA': 'DataAplic',
                        'DATAVENCIMENTO': 'DataVenc'
                    }, inplace=True)
                    
                    df_filtrado['Descricao'] = df_filtrado['TITULO'].fillna('')
                    df_filtrado.drop(columns=['TITULO'], inplace=True)
                    
                    df_filtrado['Qnt'] = df_filtrado['QUANTIDADE']
                    df_filtrado.drop(columns=['QUANTIDADE'], inplace=True)
                    
                    df_01_renda_fixa = pd.concat([df_01_renda_fixa, df_filtrado], ignore_index=True)
            
            elif '02-Fundos-Fundos' in pasta:
                colunas_relevantes = ['CARTEIRA', 'DATA', 'VALORLIQUIDO', 'QUANTIDADEDISPONIVEL', 'FUNDO', 'CODIGO']
                
                if all(col in df.columns for col in colunas_relevantes):
                    df_filtrado = df[colunas_relevantes]
                    
                    # Renomeando colunas
                    df_filtrado.rename(columns={
                        'CARTEIRA': 'NmFundo',
                        'DATA': 'DtPosicao',
                        'VALORLIQUIDO': 'VlrMercado'
                    }, inplace=True)
                    
                    # Extraindo o texto completo para identificar melhor
                    # Extrair "A VENCER" ou "VENCIDOS" de forma mais precisa
                    df_filtrado['TipoCredito'] = df_filtrado['FUNDO'].apply(
                        lambda x: 'VENCIDOS' if 'VENCIDOS' in str(x) or 'VENCIDO' in str(x) else 
                                ('A VENCER' if 'A VENCER' in str(x) or 'A VENCE' in str(x) or 'A VENC' in str(x) else None)
                    )
                    
                    # Usar a descrição extraída
                    df_filtrado['Descricao'] = df_filtrado['TipoCredito']
                    df_filtrado.drop(columns=['TipoCredito'], inplace=True)
                    
                    # Definindo a Categoria como "DIREITOS CREDITÓRIOS"
                    df_filtrado['Categoria'] = df_filtrado['Descricao'].apply(
                        lambda x: "DIREITOS CREDITORIOS" if isinstance(x, str) and (x == 'A VENCER' or x == 'VENCIDOS') else ""
                    )
                    
                    # Ajustando a quantidade
                    df_filtrado['Qnt'] = df_filtrado['QUANTIDADEDISPONIVEL']
                    df_filtrado.drop(columns=['QUANTIDADEDISPONIVEL'], inplace=True)
                    
                    # Concatenando os dados tratados
                    df_02_fundos = pd.concat([df_02_fundos, df_filtrado], ignore_index=True)

            
            elif '03-Outros_Ativos' in pasta:
                colunas_relevantes_outros = ['CARTEIRA', 'DATA', 'VALOR', 'ATIVO']
                if all(col in df.columns for col in colunas_relevantes_outros):
                    df_filtrado = df[colunas_relevantes_outros]
                    df_filtrado = df_filtrado[df_filtrado['ATIVO'].str.contains('PDD/MTM', na=False, case=False)]
                    df_filtrado = df_filtrado[['CARTEIRA', 'DATA', 'VALOR']]
                    df_filtrado['VALOR'] = (
                        df_filtrado['VALOR']
                        .str.replace('.', '', regex=False)
                        .str.replace(',', '.', regex=False)
                        .astype(float)
                    )
                    df_filtrado.rename(columns={
                        'CARTEIRA': 'NmFundo',
                        'DATA': 'DtPosicao',
                        'VALOR': 'VlrMercado'
                    }, inplace=True)
                    df_filtrado['Descricao'] = None
                    df_filtrado['Qnt'] = None
                    df_outros_ativos = pd.concat([df_outros_ativos, df_filtrado], ignore_index=True)



# Adicionar coluna "Fonte" para identificar a origem dos dados
df_06_patrimonio['Fonte'] = '06-Patrimonio-Totais'
df_01_renda_fixa['Fonte'] = '01-Renda-Fixa'
df_02_fundos['Fonte'] = '02-Fundos-Fundos'
df_outros_ativos['Fonte'] = '03-Outros-Ativos'

# Concatenar os DataFrames verticalmente, preenchendo com NULL (NaN)
df_master = pd.concat([df_06_patrimonio, df_01_renda_fixa, df_02_fundos, df_outros_ativos], ignore_index=True)

#Pegando o tipo de fundo a partir de NmFundo
df_master['Tp_Fundo'] = df_master['NmFundo'].str.split().str[0]


#Coluna Grupo
df_master['Grupo'] = df_master['Fonte'].map(grupo_map)

#Incluindo a Categoria
df_master['Categoria'] = df_master["Grupo"].map(descricao_dict2)

# Verificar se a Descricao contém "A VENCER" ou "VENCIDOS" e alterar as colunas Grupo e Categoria
df_master.loc[df_master['Descricao'].str.contains('A VENCER|VENCIDOS', na=False), 'Categoria'] = "DIREITOS CREDITORIOS"
df_master.loc[df_master['Descricao'].str.contains('A VENCER|VENCIDOS', na=False), 'Grupo'] = "DIREITOS CREDITORIOS"

# Verificar se a Descricao contém "A VENCER" ou "VENCIDOS" e alterar as colunas Grupo e Categoria
df_master.loc[df_master['Descricao'].str.contains('A VENCER|VENCIDOS', na=False), 'Categoria'] = "DIREITOS CREDITORIOS"
df_master.loc[df_master['Descricao'].str.contains('A VENCER|VENCIDOS', na=False), 'Grupo'] = "DIREITOS CREDITORIOS"

# Adicionar alerta caso não haja direitos creditórios
nao_creditorios = df_master[df_master['Categoria'] != "DIREITOS CREDITORIOS"]
if not nao_creditorios.empty:
    print("Alerta: Existem registros sem a categoria de 'DIREITOS CREDITORIOS'. Verifique os dados.")
    print(nao_creditorios[['NmFundo', 'Descricao', 'Categoria']])  # Exibe os fundos e suas descrições

df_master['VlrMercado'] = df_master['VlrMercado'].apply(ajustar_valor) 

df_master = df_master.apply(atualizar_grupo, axis=1)

df_master = df_master.apply(atualizar_desc, axis=1)

#Coluna Cod
df_master['Cod'] = df_master['Grupo'].map(grupo_cod)

#inclui a coluna 'Origem'
df_master['Custodiante'] = 'Master'

# Definir a nova ordem das colunas
nova_ordem = ['NmFundo', 'Tp_Fundo', 'DtPosicao', 'Descricao', 'Qnt', 'VlrMercado', 'DataVenc', 'DataAplic', 'Categoria', 'Cod', 'Grupo', 'Custodiante']

# Reorganizar as colunas no DataFrame
df_master = df_master[nova_ordem]

df_master['DtPosicao'] =  pd.to_datetime(df_master['DtPosicao'], dayfirst=True, errors='coerce').dt.strftime('%Y-%m-%d')

# Exibir as primeiras linhas do DataFrame combinado
df_master.head(50)
```
